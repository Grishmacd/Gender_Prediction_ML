```python

from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# Image name (upload this image)
IMAGE_NAME = "person.jpg"   # change if needed

# Load image
img = cv2.imread(IMAGE_NAME)

if img is None:
    raise ValueError("‚ùå Image not found. Upload person.jpg")

# Analyze gender
result = DeepFace.analyze(
    img_path=IMAGE_NAME,
    actions=["gender"],
    enforce_detection=False
)

# Extract gender
gender = result[0]["dominant_gender"]

print("üéØ Predicted Gender:", gender)

# Show image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title(f"Predicted Gender: {gender}")
plt.axis("off")
plt.show()
```
